```python
class Dog:
    legs = 4
    bread = 'pitbull'
    action = 'bite', 'jump'
    
    def __init__(self, name, age):
        print('creating a dog object')
        self.name = name
        self.age = age
    
    def bark(self, anger_level=1):
        print('bhoo bhoo'*anger_level)
    
    def run(self, fear_level=1):
        print('run run'*fear_level)
    
```

```python
d1 = Dog('mack', 2)
d2 = Dog('jack', 1)
```

```python
print(d1.bread, d1.legs, d1.action , d1.name, d1.age)
print(d2.bread, d2.legs, d2.action, d2.name, d2.age)
```

```python
class Counter:
    def __init__(self, x=0, step=1):
        self.step = step
        self.x = x
    def up(self):
        self.x += self.step
    def down(self):
        self.x -= self.step
    def get_current(self):
        return self.x
```

```python
c1 = Counter(4,5)
```

```python
c1.get_current()
```

```python
class Coordinate:
    def __init__(self, x = 0, y=0):
        self.x = x
        self.y = y
    def getx(self):
        return self.x
    def gety(self):
        return self.y
    def distance(self, other):
        x_diff_sq = (self.x - other.x)**2
        y_diff_sq = (self.y - other.y)**2
        return (x_diff_sq + y_diff_sq)**0.5
    def __str__(self):
        return '(' + str(self.x) + ',' + str(self.y) + ')'
    def __add__(self, other):
        return '(' + str(self.x + other.x) + ',' + str(self.y + other.y) + ')'
```

```python
c1 = Coordinate()
```

```python
c1.getx()
```

```python
c1.gety()
```

```python
c2 = Coordinate(3,4)
```

```python
c2.getx()
```

```python
c2.gety()
```

```python
c1.distance(c2)
```

```python
class Stack:
    def __init__(self):
        self.__x = []
    def push(self, a):
        self.x.append(a)
    def pop(self):
        return self.x.pop()
    def top(self):
        return self.x[-1]
    def is_empty(self):
        return len(self.x) == 0
```

```python
s1 = Stack()
```

```python
type(s1)
```
